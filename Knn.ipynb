```python
import pandas as pd
df=pd.read_csv("D:/data science/Assignments/Knn/Zoo.csv")

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["animal name"]=le.fit_transform(df["animal name"])
Y=df["type"]
X=df.iloc[:,1:17]

from sklearn.preprocessing import MinMaxScaler
MM=MinMaxScaler()
MM_X=MM.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(MM_X,Y,test_size=0.3)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=21,p=2)
knn.fit(X_train,Y_train)
Y_pred_train=knn.predict(X_train)
Y_pred_test=knn.predict(X_test)
from sklearn.metrics import accuracy_score
ac1=accuracy_score(Y_train,Y_pred_train)
ac2=accuracy_score(Y_test,Y_pred_test)
print("Training accuracy",ac1.round(3))
print("Testing accuracy",ac2.round(3))
```
