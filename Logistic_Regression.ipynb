```python
X=df.iloc[:,1:16]
Y=df["y"]
print(X)
```

```python
Y
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss_count=ss.fit_transform(X)
ss_count
ss_count=pd.DataFrame(ss_count)
ss_count
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(ss_count,Y)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
Y_pred_train=lr.predict(X_train)
Y_pred_test=lr.predict(X_test)
print(Y_pred_train)
Y_pred_test
Y_pred_test=pd.DataFrame(Y_pred_test)
Y_pred_test
```

```python
from sklearn.metrics import accuracy_score,f1_score
ac1=accuracy_score(Y_train,Y_pred_train)
ac2=accuracy_score(Y_test,Y_pred_test)
f1=f1_score(Y_train,Y_pred_train)
print(f1.round(2))
print("Training accuracy",ac1)
print("Test accuracy",ac2)
```

```python
train_error = []
test_error = []

for i in range(1,50):
    X_train,X_test,Y_train,Y_test = train_test_split(ss_count,Y, test_size=0.3, random_state=(i))
    LR = LogisticRegression()
    LR.fit(X_train,Y_train)
    Y_pred_train = LR.predict(X_train)
    Y_pred_test = LR.predict(X_test)
    train_error.append(accuracy_score(Y_train,Y_pred_train))
    test_error.append(accuracy_score(Y_test,Y_pred_test))
    
import numpy as np    
print('Training MSE - CV: ',(np.mean(train_error).round(2)))
print('Test MSE - CV: ',(np.mean(test_error).round(2)))
```

```python
Y_proba = lr.predict_proba(ss_count)[:,1]
df["Y_proba"] = pd.DataFrame(Y_proba)
df.head()
```

```python
def f1(X):
    if X < 0.5:
        return 0
    elif X >= 0.5:
        return 1

df['Class_predicted'] = df["Y_proba"].apply(f1)

pd.crosstab(df["y"],df["Class_predicted"])
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
fpr, tpr,threshold  = roc_curve(Y,Y_proba)

plt.plot(fpr,tpr)
plt.ylabel('tpr - True Positive Rate')
plt.xlabel('fpr - False Positive Rate')
plt.show()
auc = roc_auc_score(Y, Y_proba)
auc.round(3)
```
