# 1) Delivery time - predicting delivery time using sorting time.

```python
import pandas as pd
import numpy as np
df=pd.read_csv("D:/data science/Assignments/Simple Linear Regression/delivery_time.csv")

df.isnull().sum()

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
plt.scatter(x=df["Sorting Time"],y=df["Delivery Time"])
plt.show()

df.corr()

df.boxplot(column="Sorting Time")

x=df[["Sorting Time"]]
y=df["Delivery Time"]

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
lr.intercept_

y_pred=lr.predict(x)

plt.scatter(x=df["Sorting Time"],y=df["Delivery Time"])
plt.plot(df["Sorting Time"],y_pred,color="k")
plt.show()

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred)
print("Mean squared error: ",mse.round(3))
print("root mean squared error:",np.sqrt(mse).round(3))


import seaborn as sb
sb.distplot(df["Sorting Time"])
fig = plt.figure()

print("\nLog Transformation")
L=[]
import seaborn as sb
L=pd.DataFrame(L)
L["Sorting Time"]=np.log(df["Sorting Time"])
sb.distplot(L["Sorting Time"])
fig = plt.figure()

x=L[["Sorting Time"]]
lr=LinearRegression()
lr.fit(x,y)
lr.intercept_

y_pred=lr.predict(x)

plt.scatter(x=L["Sorting Time"],y=df["Delivery Time"])
plt.plot(L["Sorting Time"],y_pred,color="k")
plt.show()

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred)
print("Mean squared error: ",mse.round(3))
print("root mean squared error:",np.sqrt(mse).round(3))


print("\nSquare Root Transformation")
sr=[]
import seaborn as sb
sr=pd.DataFrame(sr)
sr["Sorting Time"]=np.sqrt(df["Sorting Time"])
sb.distplot(sr["Sorting Time"])
fig = plt.figure()

x=sr[["Sorting Time"]]
lr=LinearRegression()
lr.fit(x,y)
lr.intercept_

y_pred=lr.predict(x)

plt.scatter(x=sr["Sorting Time"],y=df["Delivery Time"])
plt.plot(sr["Sorting Time"],y_pred,color="k")
plt.show()

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred)
print("Mean squared error: ",mse.round(3))
print("root mean squared error:",np.sqrt(mse).round(3))



print("\nCubic Root Transformation")
c=[]
import seaborn as sb
c=pd.DataFrame(c)
c["Sorting Time"]=np.log(df["Sorting Time"])
sb.distplot(c["Sorting Time"])
fig = plt.figure()

x=c[["Sorting Time"]]
lr=LinearRegression()
lr.fit(x,y)
lr.intercept_

y_pred=lr.predict(x)

plt.scatter(x=c["Sorting Time"],y=df["Delivery Time"])
plt.plot(c["Sorting Time"],y_pred,color="k")
plt.show()

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred)
print("Mean squared error: ",mse.round(3))
print("root mean squared error:",np.sqrt(mse).round(3))

print("\nSquare Transformation")
s=[]
import seaborn as sb
s=pd.DataFrame(s)
s["Sorting Time"]=np.power(df["Sorting Time"],2)
sb.distplot(s["Sorting Time"])
fig = plt.figure()

x=s[["Sorting Time"]]
lr=LinearRegression()
lr.fit(x,y)
lr.intercept_

y_pred=lr.predict(x)

plt.scatter(x=s["Sorting Time"],y=df["Delivery Time"])
plt.plot(s["Sorting Time"],y_pred,color="k")
plt.show()

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,y_pred)
print("Mean squared error: ",mse.round(3))
print("root mean squared error:",np.sqrt(mse).round(3))
```
